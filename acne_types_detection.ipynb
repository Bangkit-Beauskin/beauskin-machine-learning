```python
from ultralytics import YOLO
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Load the trained model
model = YOLO('/content/drive/MyDrive/BEAUSKIN ML/models/best_model.pt')  # path to your best model

# Make a prediction on a sample image
results = model.predict("/content/drive/MyDrive/BEAUSKIN ML/ACNE_DETECTION/train/Acne/16.jpg")

# Get the image and predictions
img_path = "/content/drive/MyDrive/BEAUSKIN ML/ACNE_DETECTION/train/Acne/16.jpg"
img = cv2.imread(img_path)  # Read image using OpenCV
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Convert to RGB for displaying in matplotlib

# Define fixed colors for each label (6 labels)
# Assign a specific color to each acne class (adjust according to your class labels)
fixed_colors = {
    0: (255, 0, 0),      # Red for 'blackheads'
    1: (0, 255, 0),      # Green for 'dark spots'
    2: (0, 0, 255),      # Blue for 'nodules'
    3: (255, 255, 0),    # Cyan for 'papules'
    4: (255, 0, 255),    # Magenta for 'pustules'
    5: (0, 255, 255),    # Yellow for 'whiteheads'
}

# Plot the image with bounding boxes
plt.figure(figsize=(5, 5))
plt.imshow(img_rgb)
plt.axis('off')

# Draw bounding boxes on the image with fixed colors for each label
for box in results[0].boxes:
    x1, y1, x2, y2 = box.xyxy[0]  # Get box coordinates
    label = int(box.cls[0].item())  # Get the class label and convert to int
    conf = box.conf[0].item()  # Get the confidence score and convert to float

    # Get a fixed color for the label
    color = fixed_colors.get(label, (255, 255, 255))  # Default to white if label not found

    # Draw the bounding box and label with the color
    cv2.rectangle(img_rgb, (int(x1), int(y1)), (int(x2), int(y2)), color, 1)  # Draw rectangle
    cv2.putText(img_rgb, f'{model.names[label]}', (int(x1), int(y1)-10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, color, 1)

# Show the image with predictions
plt.imshow(img_rgb)
plt.axis('off')
plt.show()
```
